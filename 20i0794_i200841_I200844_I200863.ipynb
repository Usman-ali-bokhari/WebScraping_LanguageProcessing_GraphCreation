```python
import requests
import re
from bs4 import BeautifulSoup
import spacy
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
```

```python
def getdata(p,tags):
    q=[]
    #Gets the HTML document 
    response = requests.get(p)
    url2=response.text
        
    #Gets Html of website
    WebHtml = BeautifulSoup(url2, "html.parser")
    #Extracts text from the tag we use to extract text
    TagHtml = WebHtml.find_all(tags)
    
    var=[]
    #converts TagHtml into text and stores in list var.
    for i in TagHtml:
        sample=i.text
        sample = sample.replace("\n","")
        sample = sample.replace("\r","")
        sample = sample.replace("\t","") 
        sample = sample.replace("\xa0","")
        sample = sample.replace("  ","") 
        var.append(sample)
    while("" in var) :
            var.remove('')
    #Concatenates var into string and stores into WebHtml.
    WebHtml ="".join(str(var)) 
    
    return WebHtml
```

```python
pd.set_option("max_rows", 400)
pd.set_option("max_colwidth", 400)
nlp = spacy.load('en_core_web_sm')
```

```python
def getdata(p, tags):
    q = []
    # Gets the HTML document
    response = requests.get(p)
    url2 = response.text

    # Gets Html of website
    WebHtml = BeautifulSoup(url2, "html.parser")
    # Extracts text from the tag we use to extract text
    TagHtml = WebHtml.find_all(tags)

    var = []
    # converts TagHtml into text and stores in list var.
    for i in TagHtml:
        sample = i.text
        sample = sample.replace("\n", "")
        sample = sample.replace("\r", "")
        sample = sample.replace("\t", "")
        sample = sample.replace("\xa0", "")
        # sample = re.sub(r"[-()\'\"#%/@;:<>{}`+=~|.!?,]", '', sample)
        var.append(sample)
    # Concatenates var intp string and stores into WebHtml.
    WebHtml = "".join(str(var))
    return WebHtml
```

```python
def degree_nouns(text):
    nouns = []
    for token in text:
        if token.pos_ == 'NOUN':
            nouns.append(token.text.lower())

        nouns_tally = Counter(nouns)
        df_noun = pd.DataFrame(nouns_tally.most_common(), columns=['noun', 'degree'])

    df_noun = pd.DataFrame(nouns_tally.most_common(10), columns=['noun', 'degree'])
    #print(df_noun)
    return df_noun
```

```python
def function_graph(text,str1,str2,str3):
    #finding adjectives
        adjs = []
        for token in text:
            if token.pos_ == 'ADJ':
                adjs.append(token.text.lower())

        adjs_tally = Counter(adjs)
        df_adj = pd.DataFrame(adjs_tally.most_common(), columns=['adj', 'count'])
        # print(df_adj)
        # print(adjs_tally.most_common())

        # finding nouns
        nouns = []
        for token in text:
            if token.pos_ == 'NOUN':
                nouns.append(token.text.lower())

        nouns_tally = Counter(nouns)
        df_noun = pd.DataFrame(nouns_tally.most_common(), columns=['noun', 'count'])
        # print(df_noun)
        # print(nouns_tally.most_common())

        # finding verbs
        verbs = []
        for token in text:
            if token.pos_ == 'VERB':
                verbs.append(token.text.lower())
        verbs_tally = Counter(verbs)
        df_verb = pd.DataFrame(verbs_tally.most_common(), columns=['verb', 'count'])
        # print(df_verb)
        # print(verbs_tally.most_common())


        # making plot for 10 top adjectives
        df_adj = pd.DataFrame(adjs_tally.most_common(10), columns=['adj', 'count'])
        # print(df_adj)
        plt.bar(df_adj['adj'], df_adj['count'], color='blue')
        plt.title(str1)
        plt.xlabel('Adjective', fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.grid(False)
        plt.xticks(rotation=25)
        plt.show()


        # making plot for 10 top nouns
        df_noun = pd.DataFrame(nouns_tally.most_common(10), columns=['noun', 'count'])
        # print(df_noun)
        plt.bar(df_noun['noun'], df_noun['count'], color='pink')
        plt.title(str2)
        plt.xlabel('Noun', fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        plt.grid(False)
        plt.xticks(rotation=25)
        plt.show()

         # making plot for 10 top verbs
        df_verb = pd.DataFrame(verbs_tally.most_common(10), columns=['verb', 'count'])
        # print(df_verb)
        plt.bar(df_verb['verb'], df_verb['count'], color='black',)
        plt.title(str3)
        plt.xlabel('Verb', fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.grid(False)
        plt.xticks( rotation=25)
        plt.show()
        return
```

```python
text_FAST=nlp(FASTText)
function_graph(text_FAST,"Top 10 Adjectives for FAST University",
         "Top 10 Nouns for FAST University",
         "Top 10 Verbs for FAST University")

text_LUMS=nlp(LUMSText)
function_graph(text_LUMS,"Top 10 Adjectives for LUMS University",
         "Top 10 Nouns for LUMS University",
         "Top 10 Verbs for LUMS University")


text_Comsats=nlp(ComsatsText)
function_graph(text_Comsats,"Top 10 Adjectives for Comsats University",
         "Top 10 Nouns for Comsats University",
         "Top 10 Verbs for Comsats University")
```
